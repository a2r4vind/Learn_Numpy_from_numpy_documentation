## 2.3 Slicing in numpy array 
Slicing in numpy array is different from slicing of python list :
    
    Differences:
        1) slicing in numpy array creates a **view** of the portion of the original array sliced 
        not a copy which is found in python list slicing.
        2) slicing doesn't create copy.
        3) changes made to the **view** reflects the changes in the original numpy array.
        

```python
a[:3]
```

```python
b = a[3:] 
b # b is a View
```

```python
b[0] = 40 # changes in view 'b' make changes in original numpy array 'a'
```

```python
a 
```

# Adding, removing, and sorting elements
1) np.sort() => can also specify **axis, kind, and order** when calling the function. Returns sorted copy of an array.


2) np.concatenate() => used to concatenate arrays.

```python
a = np.array([2, 1, 5, 3, 7, 4, 6, 8])
print(a)
```

```python
# sorting in ascending order.
a_sorted = np.sort(a)
print(a_sorted)
```

**argsort** => which is an indirect sort along a specified axis,

**lexsort** => which is an indirect stable sort on multiple keys,

**searchsorted** => which will find elements in a sorted array, 

**partition** => which is a partial sort.

```python
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
```

```python
concatenated_arrys = np.concatenate((x,y))
print(concatenated_arrys)
```

```python
v1 = np.array([[1,2],[3,4]])
v2 = np.array([[5, 6]])

v_prime = np.concatenate((v1,v2), axis=0)
print(v_prime)
print()
print(f'shape: {v_prime.shape}')
```
