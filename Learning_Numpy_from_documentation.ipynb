## 2.3 Slicing in numpy array 
Slicing in numpy array is different from slicing of python list :
    
    Differences:
        1) slicing in numpy array creates a **view** of the portion of the original array sliced 
        not a copy which is found in python list slicing.
        2) slicing doesn't create copy.
        3) changes made to the **view** reflects the changes in the original numpy array.
        

```python
a[:3]
```

```python
b = a[3:] 
b # b is a View
```

```python
b[0] = 40 # changes in view 'b' make changes in original numpy array 'a'
```

```python
a 
```

**Custom filtering**

```python
# to get all values below 8
below_eight = b < 8
print(b[below_eight])
```

```python
# get number equal to or greater than 5
five_up = b >= 5
print(b[five_up])
```

```python
# get elements divisible by 2
divisible_by_2 = b[b%2==0]
divisible_by_2
```

***Note:-** to combine 2 or more conditions use **&** and **|**

```python
# get numbers greater than 2 but less than or equal 11
required_numbers = b[(b > 2) & (b <= 11)]
required_numbers
```

**Note:-** **&** and **|** can also be used to get boolean arrays

```python
# get an array of boolean values that satisfy the following conditions are values greater than or equal to 5?
five_up_bool = (b > 5) | (b == 5)
print(five_up_bool)
```

**Note:-** **np.nonzero()** can be used to select elements or indices from an array.

```python
# get the indices of the elements of b that are less than 5
c = np.nonzero(b < 5)
c
```

1) Returns the output as tuple of arrays: one for each dimension. 
2) Here, first array represents the row indices where the value is found 
and the second array represents the columns indices where the value is found.

```python
# printing the above selected elements
print(b[c])
```

**Note:-** if the value is not present then it will give an empty array

**.view()** method gives a shallow copy of the data

slicing and indexing also returns a **view** of the original data.

```python
v = np.arange(1,13)
v
```

```python
z = v.view()
z
```

```python
# or simply
y = v
y
```

```python
# change 4 at index 3 using z
print(z[3])
z[3] = 44
print(z)
```

```python
# now print original vector v
v
```

```python
# change 9 at index 8 using y
print(y[8])
y[8] = 99
y
```

```python
# now print original vector v
v
```

**.copy()** method returns a deep copy of the data

```python
v1 = np.arange(1,15)
v1
```

```python
v1_cp = v1.copy()
v1_cp
```

```python
# change value 4 at index 3 in v1_cp
print(v1_cp[3])
v1_cp[3] = 44
v1_cp
```

```python
# now print the original vector v1
v1
```
